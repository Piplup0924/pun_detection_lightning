```python
import re
```

```python
def clean_up(x):
    emojipat = re.compile("\[.*?\]")
    apat = re.compile("\{\{\w*_\d*\}\}")
    spanpat = re.compile("<.*?>")
    tagpat = re.compile("#.*?#")
    # x = x.replace("<p>","").replace("</p>", "")
    x = emojipat.sub("", x)
    x = tagpat.sub("", x)
    x = apat.sub("", x)
    x = spanpat.sub("", x)
    return x.strip()
```
